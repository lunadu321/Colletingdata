```python
#-------------------------------------------------------------------------------------------------------------------------#
#
#                     MOVIES INDUSTRY PROJECT - EXPLORATORY DATA ANALYSIS AND DATA VISUALIZATION
#
#-------------------------------------------------------------------------------------------------------------------------#


import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Reading in the data and taking a glimpse at it

df = pd.read_csv('C:\\Users\\dudud\\collecting data\\movie_statistic_dataset.csv')

df.head()
```

```python
# There's no column of the released year in the column.
# Therefore I create a new column of year information extracted from the production date.

df['year'] = ((df['production_date'].str.extract(pat = '([0-9]{4})'))[0]).str.strip()
df['year'] = df['year'].astype('int64')

df.head()
```

```python
# Checking columns' data types

df.dtypes
```

```python
# Descriptive statistics for the entire dataset

df.describe(include='all')
```

```python
#Some insights:
# This dataset has 4380 titles.
# The studied year lapse goes from 1915 to 2023.
# The average film duration is 1h 49min.
# The most common genre is comedy,drama,romance.
```

```python
# Sorting data by gross revenue in a descending order, showing the top 10 grossing movies.

df.sort_values(by = ['Worldwide gross $'], ascending = False).head(10)
```

```python
# Visualizing 'budget' vs 'gross' in a scatter plot.

plt.figure(figsize=(12, 8))
plt.scatter(df['Production budget $'], df['Worldwide gross $'], alpha=0.5)
plt.title('Budget vs Worldwide Gross')
plt.xlabel('Budget')
plt.ylabel('Worldwide Gross')
plt.show()
```

```python
# Visualizing 'budget' vs 'gross' in a regression plot.

plt.figure(figsize=(12, 8))
sns.regplot(x='Production budget $', y='Worldwide gross $', data=df, scatter_kws={'alpha':0.5})
plt.title('Budget vs Worldwide Gross (Regression Plot)')
plt.xlabel('Budget')
plt.ylabel('Worldwide Gross')
plt.show()
```

```python
# INSIGHT: The scatter plot suggests there's some positive correlation between the 2 variables.
#          The regression plot confirms the positive correlation.
#          Movies with higher production budget are likely to generate higher revenue.
```

```python
# Movies released over the years

movie_counts_per_year = df['year'].value_counts().sort_index()
plt.figure(figsize=(12, 8))
plt.bar(movie_counts_per_year.index, movie_counts_per_year, color='orange')
plt.title('Number of Movies Released Over the Years')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies Released')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# INSIGHT: Starting from 1980s, movies industry had grown significantly until the early 2000s, then stabilized.
#          The drop at the end of the plot is due to the lack of data in recent years.
```

```python
# To find the top five years with most released films
top_years = df['year'].value_counts().head(5)

# Plotting the top five years
plt.figure(figsize=(10, 6))
top_years.plot(kind='bar', color='blue')
plt.title('Top 5 Years with Most Released Films')
plt.xlabel('Release Year')
plt.ylabel('Number of Films Released')
plt.show()
```
